# SelectKBest based on mutual information - SKlearn

```python
# get_support to maintain the index name
def function_K_selector(features,labels, function, k_features):
    selector = SelectKBest(function, k_features).fit(features,labels)
    return features[features.columns[selector.get_support(indices=True)]], selector.scores_
```

```python
# why 40 features?
X_collinear_MI, scores = function_K_selector(X_collinear, y, mutual_info_regression, 40)
X_collinear_MI.shape
```

```python
pd.DataFrame(X_collinear_MI).head()
```

```python
pd.DataFrame(scores).sort_values(by=0, ascending=False).head()
```

```python
X_collinear_MI_score = pd.DataFrame({'Feature':list(pd.DataFrame(X_collinear).columns),
                                     'Scores':scores}).set_index('Feature') #set feature columns as index
X_collinear_MI_score.sort_values(by='Scores', ascending=False)#.reset_index(drop=True)
# The feature labels are tramsformed to nemeric index due to the operation of VarianceThreshold
```

```python
# plot 30 features with the largets scores
# seasonal_num is the seasonal schedule
fig = X_collinear_MI_score.nlargest(30, columns='Scores').sort_values('Scores', ascending=True).plot.barh(
    figsize=(10,8), use_index=True)
fig.set_ylabel("Features")
fig.set_xlabel("Scores")
plt.show()
```

# Trial on selecting k

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
```

```python
clf = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
```

```python
# search for an optimal value of k for RandomForrestRegressor
k_range = list(range(60, 71))
k_scores = []
for k in k_range:
    X_collinear_MI, scores = function_K_selector(X_collinear, y, mutual_info_regression, k)
    scores_model = cross_val_score(clf, X_collinear_MI, y, cv=5)
    k_scores.append(scores_model.mean())
print(k_scores)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# plot the value of K for KNN (x-axis) versus the cross-validated accuracy (y-axis)
plt.plot(k_range, k_scores)
plt.xlabel('Value of K features')
plt.ylabel('Cross-Validated Accuracy')
```
