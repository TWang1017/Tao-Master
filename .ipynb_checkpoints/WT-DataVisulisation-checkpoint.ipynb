```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from sklearn.preprocessing import normalize
import seaborn as sns
import matplotlib
```

```python
# list of models
# Commented few models because they produced very big results which interfere visualization
models = [
          'ExtraTreesRegressor',
        'ExtraTreeRegressor',
        'RandomForestRegressor',
          'BaggingRegressor',
          'DecisionTreeRegressor',
                   'MLPRegressor',
          
          
           #'GaussianProcessRegressor',
          'GradientBoostingRegressor',
          'HuberRegressor',
    'TheilSenRegressor',
    'AdaBoostRegressor',
          'KNeighborsRegressor',
       'DummyRegressor'

          #'PassiveAggressiveRegressor',
          #'RANSACRegressor',
          #'SGDRegressor',
          
    
  
         ]
buildingtypes = ['Office', 'PrimClass', 'UnivClass', 'UnivDorm', 'UnivLab']
```

```python
def assign_buildingtype(name):
    if 'Office' in name:
        return 'Office'
    elif "PrimClass" in name:
        return "PrimClass"
    elif "UnivClass" in name:
        return "UnivClass"
    elif "UnivDorm" in name:
        return "UnivDorm"
    else:
        return "UnivLab"
```

```python
f, axes = plt.subplots(5,3, figsize=(20,40), sharex= 'col', sharey='row')
plt.style.use('seaborn-whitegrid')
for types in buildingtypes: #loop different building types rowwise in different plots
    dataframes = []
    for i in range(len(models)): #loop the ml models in one plot
        cv_range = range(1,5)
        for cv_mode in cv_range: #loop the TimeSeriesSplit cv in one plot
            data = pd.read_csv('/Users/t.wang/Documents/GitHub/building-prediction-benchmarking/results/' + 
                               models[i] + '_metrics_cross_validation_' + str(cv_mode) + '.csv')
            data['buildingtype'] = data.building.apply(lambda x: assign_buildingtype(x))
            data['cv_mode'] = str(cv_mode)
            data['model'] = models[i]
            dataframes.append(data)
            result = pd.concat(dataframes)

    MAPE = result.melt(id_vars=['buildingtype','cv_mode','model'], value_vars='MAPE')
    ax1 = sns.boxplot(data=MAPE[MAPE.buildingtype == types][MAPE<100], 
                      x='value', y='model', hue='cv_mode', ax=axes[buildingtypes.index(types),0])
    ax1.set(ylabel=types, xlabel="MAPE") #in one row, one building type, place MAPE at first column
    
    NMBE = result.melt(id_vars=['buildingtype','cv_mode','model'], value_vars='NMBE')
    ax2 = sns.boxplot(data=NMBE[NMBE.buildingtype == types][(NMBE<50)&(NMBE>-50)], 
                      x='value', y='model', hue='cv_mode', ax=axes[buildingtypes.index(types),1])
    ax2.set(ylabel='', xlabel="NMBE") #in one row, one building type, place NMBE at second column
    
    CVRSME = result.melt(id_vars=['buildingtype','cv_mode','model'], value_vars='CVRSME')
    ax3 = sns.boxplot(data=CVRSME[CVRSME.buildingtype == types][CVRSME<100], 
                      x='value', y='model', hue='cv_mode', ax=axes[buildingtypes.index(types),2])
    ax3.set(ylabel='', xlabel="CVRSME") #in one row, one building type, place CVRSME at third column
    
# for one metric, loop one ml model with different cv, then loop for all ml models, all happen in one plot
# then loop different metrics, placing them rowwise
# then loop different buildingtypes, placing them columnwise

```

```python
name_mapping = {"Office":"Office","PrimClass":"Primary School","UnivDorm":"Univ. Dorm","UnivClass":
                "Univ. Classroom","UnivLab":"Univ. Lab"}
color_palattes = ["viridis","spring","summer","autumn","winter"]
```

```python
for types in buildingtypes: 
    f, axes = plt.subplots(1,4, figsize=(12,6))#, sharex= 'col', sharey='row')
    #instead of treating the the graph as a whole, we treat each building type 
    #as one and then loop 5 times (5buildingtypes), then overlay each row
    plt.style.use('seaborn-whitegrid')
    f.suptitle(name_mapping[types]+" Model Comparisons", fontsize=20)
    
    models_in_order = list(result[result.buildingtype == 
       'Office'].groupby('model').median().abs().drop('RSQUARED', axis=1).sum(axis=1).sort_values().index)
    
    dataframes = []
    
    for i in range(len(models_in_order)): 
        cv_range = range(1,5)
        for cv_mode in cv_range: 
            data = pd.read_csv('/Users/t.wang/Documents/GitHub/building-prediction-benchmarking/results/' + 
                               models[i] + '_metrics_cross_validation_' + str(cv_mode) + '.csv')
            data['buildingtype'] = data.building.apply(lambda x: assign_buildingtype(x))
            data['cv_mode'] = str(cv_mode)
            data['model'] = models_in_order[i] #use the model index in order
            dataframes.append(data)
            result = pd.concat(dataframes)

    MAPE = result.melt(id_vars=['buildingtype','cv_mode','model'], value_vars='MAPE')
    ax1 = sns.boxplot(data=MAPE[MAPE.buildingtype == types][MAPE<100], 
                      x='value', y='model', hue='cv_mode', palette=color_palattes[buildingtypes.index(types)], ax=axes[0])
    ax1.set(ylabel='', xlabel="MAPE") 
    ax1.get_legend().remove()
    
    NMBE = result.melt(id_vars=['buildingtype','cv_mode','model'], value_vars='NMBE')
    ax2 = sns.boxplot(data=NMBE[NMBE.buildingtype == types][(NMBE<50)&(NMBE>-50)], 
                      x='value', y='model', hue='cv_mode',palette=color_palattes[buildingtypes.index(types)], ax=axes[1])
    ax2.set(ylabel='', xlabel="NMBE",yticks=([])) 
    ax2.get_legend().remove() #yticks to disable y labels
    
    CVRSME = result.melt(id_vars=['buildingtype','cv_mode','model'], value_vars='CVRSME')
    ax3 = sns.boxplot(data=CVRSME[CVRSME.buildingtype == types][CVRSME<100], 
                      x='value', y='model', hue='cv_mode',palette=color_palattes[buildingtypes.index(types)], ax=axes[2])
    ax3.set(ylabel='', xlabel="CVRSME",yticks=([])) 
    ax3.get_legend().remove()
    
    cmap = matplotlib.cm.get_cmap(color_palattes[buildingtypes.index(types)])
    rgba = cmap(0.5)
    ranking = pd.DataFrame({"metric":result[result.buildingtype == types].groupby("model").median().abs().drop([
        "RSQUARED","NMBE"],axis=1).sum(axis=1).sort_values()}).reset_index() #two were dropped
    print(ranking)
    ax4 = sns.barplot(data=ranking, x="metric", y="model", ax=axes[3], color=rgba)
    ax4.set(ylabel="", xlabel="Cumulative Error Metric", yticks=([]))
    
    plt.show()
```

```python
myColors = ((0.8, 0.0, 0.0, 1.0), (0.0, 0.8, 0.0, 1.0), (0.0, 0.0, 0.8, 1.0))
cmap = LinearSegmentedColormap.from_list('Custom', myColors, len(myColors))
```

```python
ax = sns.heatmap(df.T, cmap=cmap, linewidths=.5, linecolor='lightgray')

# Manually specify colorbar labelling after it's been generated
colorbar = ax.collections[0].colorbar
colorbar.set_ticks([0.33, 1, 1.66]) #location of ticks
colorbar.set_ticklabels(['Not Used','Train', 'Test'])

# X - Y axis labels
ax.set_ylabel('Training/Testing Scenarios')
ax.set_xlabel('Months')

# Only y-axis labels need their rotation set, x-axis labels already have a rotation of 0
_, labels = plt.yticks()
plt.setp(labels, rotation=0)
plt.tight_layout()
```
