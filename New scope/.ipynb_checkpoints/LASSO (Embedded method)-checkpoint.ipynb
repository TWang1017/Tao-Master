```python
import pandas as pd
import numpy as np
from feature_selector import FeatureSelector
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import SelectPercentile, mutual_info_regression
import warnings
warnings.filterwarnings('ignore')
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestRegressor
import time
from sklearn.linear_model import LassoCV
%matplotlib inline
```

# LASSO feature importance selection after filter method

```python
%store -r X_collinear_Variance_MI #read the features and labels data after filter method
%store -r y
```

```python
X_collinear_Variance_MI.head()
```

```python
y.head()
```

```python
reg = LassoCV(random_state=42, n_jobs=-1, cv=10)
```

```python
start = time.time()

# Train the regressor exhaustively with all features
reg.fit(X_collinear_Variance_MI, y)

end = time.time()
elapsed = end - start 
print('Time for training all features after Filter method:'+ time.strftime("%H:%M:%S", time.gmtime(elapsed)))
```

```python
print("Best alpha using built-in LassoCV: %f" % reg.alpha_)
print("Best score using built-in LassoCV: %f" %reg.score(X_collinear_Variance_MI,y))
coef = pd.Series(reg.coef_, index = X_collinear_Variance_MI.columns)
```

```python
print("Lasso picked " + str(sum(coef != 0)) + " variables and eliminated the other " +  str(sum(coef == 0)) + " variables")
```

```python
imp_coef = coef.sort_values()
import matplotlib
matplotlib.rcParams['figure.figsize'] = (8.0, 10.0)
imp_coef.plot(kind = "barh")
plt.title("Feature importance using Lasso Model")
```

```python
num_picked = str(sum(coef != 0))
print(num_picked)
```

```python
pd.DataFrame(abs(coef).nlargest(int(num_picked)))
```

```python
picked_fea_name = pd.DataFrame(abs(coef).nlargest(int(num_picked))).index
pd.DataFrame(picked_fea_name)
```

```python
X_collinear_Variance_MI_LASSO = X_collinear_Variance_MI[picked_fea_name]
```
