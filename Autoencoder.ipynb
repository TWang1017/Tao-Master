```python
import numpy as np
from keras.datasets import mnist
from keras.models import Model
from keras.layers import Dense, Input
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
# Read one year data from a random building after FE
features = pd.read_csv('/Users/t.wang/Desktop/Dissertation/Python/WT-result/Filter_One_year_features.csv')
target = pd.read_csv('/Users/t.wang/Desktop/Dissertation/Python/WT-result/Filter_One_year_target.csv')
```

```python
features.head()
```

```python
# Data preprocessing - standardise the features
features_std = StandardScaler().fit_transform(features)
features_std = pd.DataFrame(features_std).head()
features_std.head()
```

```python
# number of input features
number_features = len(features_std.columns)
print(number_features)
```

```python
# compressed layer dimension (final feature numbers)
encoding_dim = 50

# this is our input placeholder
input_features = Input(shape=(number_features,))
# print(input_features)
```

```python
# encoder layers
encoded = Dense(1000, activation='relu')(input_features)
encoded = Dense(500, activation='relu')(encoded)
encoded = Dense(250, activation='relu')(encoded)
encoder_output = Dense(encoding_dim, name = 'bottleneck')(encoded)

# decoder layers
decoded = Dense(250, activation='relu')(encoder_output)
decoded = Dense(500, activation='relu')(decoded)
decoded = Dense(1000, activation='relu')(decoded)
decoded = Dense(number_features, activation='tanh')(decoded)
```

```python
# construct the autoencoder model
autoencoder = Model(inputs=input_features, outputs=decoded)

# construct the encoder model for dimensionality reduction
encoder = Model(inputs=input_features, outputs=encoder_output)

# compile autoencoder
autoencoder.compile(optimizer='adam', loss='mse')

epochs = 200

# training, the features are reconstructed after the entire cycle of autoencoder, therefore the input and outputs
# are both features_std
autoencoder_n = autoencoder.fit(features_std, features_std,
                epochs=epochs,
                batch_size=None,
                shuffle=True)
```

```python
# plotting
loss = autoencoder_n.history['loss']
epochs_n = range(epochs)
plt.figure()
plt.plot(epochs_n, loss, color='blue', marker='o', linestyle='solid', label='Training loss')
plt.title('Training and validation loss')
plt.xlabel('Number of features')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
# minimal loss
print('The minimal loss is ' + str(min(loss)))
# index of the min loss, +1 since the index starts from 0
index_min = loss.index(min(loss))+1
print('The index of minimal loss is ' + str(index_min))
dc_minloss = loss[index_min-1]
print('Double check the minimal loss from the index ' + str(dc_minloss))
```
